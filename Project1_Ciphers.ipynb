## Caesar Cipher Basics

```python
# Encryption
def encrypt(message, shift_amount):
    encrypted = ''
    for char in message:
        if char.isalpha():
            ascii_offset = 65 if char.isupper() else 97
            encrypted_char = chr((ord(char) - ascii_offset + shift_amount) % 26 + ascii_offset)
            encrypted += encrypted_char
        else:
            encrypted += char
    return encrypted
```

```python
# Decryption
def decrypt(message, shift_amount):
    return encrypt(message, -shift_amount)
```

```python
encrypted_text = encrypt('HELLO', 3)
print(f'Encrypted: {encrypted_text}')  # KHOOR
print(f'Decrypted: {decrypt(encrypted_text, 3)}')  # HELLO
```

## Vigen√®re Cipher

```python
def vigenere_encrypt(plaintext, keyword):
    encrypted = ''
    keyword_repeated = ''
    keyword_index = 0
    
    for char in plaintext:
        if char.isalpha():
            keyword_repeated += keyword[keyword_index % len(keyword)]
            keyword_index += 1
        else:
            keyword_repeated += char
    
    for i in range(len(plaintext)):
        if plaintext[i].isalpha():
            ascii_offset = 65 if plaintext[i].isupper() else 97
            shift = ord(keyword_repeated[i].upper()) - 65
            encrypted_char = chr((ord(plaintext[i]) - ascii_offset + shift) % 26 + ascii_offset)
            encrypted += encrypted_char
        else:
            encrypted += plaintext[i]
    
    return encrypted
```

```python
# Example
message = 'HELLO WORLD'
key = 'KEY'
encrypted = vigenere_encrypt(message, key)
print(f'Encrypted: {encrypted}')
```
